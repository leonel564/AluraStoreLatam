```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

df_combined = pd.concat([tienda, tienda2, tienda3, tienda4])
df_combined.head()
```

## Calculate total sales

### Subtask:
Calculate the total sales amount by summing the 'Precio' column from the combined dataset.


**Reasoning**:
To calculate the total sales amount, I will sum the 'Precio' column from the `df_combined` DataFrame and then print the result.



```python
total_sales = df_combined['Precio'].sum()
print(f"Total Sales: {total_sales:,.2f}")
```

## Analyze sales over time

### Subtask:
Convert the 'Fecha de Compra' column to datetime format and analyze how sales vary across different periods (e.g., year, month, day).


**Reasoning**:
To enable time-based analysis, I need to convert the 'Fecha de Compra' column to a datetime format. This will allow for the extraction of year and month components later.



```python
df_combined['Fecha de Compra'] = pd.to_datetime(df_combined['Fecha de Compra'], format='%d/%m/%Y')
df_combined.head()
```

**Reasoning**:
To analyze sales over time, I will extract the year and month from the 'Fecha de Compra' column, create new columns for them, and then group the data by these new columns to calculate the total sales ('Precio') for each year and month.



```python
df_combined['Año de Compra'] = df_combined['Fecha de Compra'].dt.year
df_combined['Mes de Compra'] = df_combined['Fecha de Compra'].dt.month

sales_by_year = df_combined.groupby('Año de Compra')['Precio'].sum()
sales_by_month = df_combined.groupby('Mes de Compra')['Precio'].sum()

print("Sales by Year:")
print(sales_by_year)
print("\nSales by Month:")
print(sales_by_month)
```

## Identify top-selling products and categories

### Subtask:
Determine which products or product categories contribute most to the total sales by grouping and summing.


**Reasoning**:
To identify the top-selling products, I will group the combined DataFrame by 'Producto' and sum the 'Precio' column, then sort the results in descending order. Similarly, for top-selling categories, I will group by 'Categoría del Producto' and sum 'Precio', also sorting in descending order. Finally, I will display the top 10 from each.



```python
top_products = df_combined.groupby('Producto')['Precio'].sum().sort_values(ascending=False)
top_categories = df_combined.groupby('Categoría del Producto')['Precio'].sum().sort_values(ascending=False)

print("Top 10 Selling Products:")
print(top_products.head(10))

print("\nTop 10 Selling Categories:")
print(top_categories.head(10))
```
